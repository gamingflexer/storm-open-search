```python
import subprocess

# List of partner names
partners = [
    "Zuber Lawler",
    "Nreach",
    "Myco",
    "Phaver",
    "Koii Network",
    "Business Inside The Game LLC",
    "Cryptonite",
    "Skale",
    "SuperWorld",
    "Alter New Media",
    "Howl",
    "Wachsman",
    "Metaverse Post",
    "Casper Labs",
    "Blockchain Training Alliance",
    "Animoca Brands",
    "DressX",
    "MTU",
    "Ae Studio",
    "Fund the Planet",
    "Startup Coil",
    "K12 Crypto",
    "Crypto Oracle",
    "Wire Network",
    "Mocaverse",
    "Gala Games"
]

# Function to run the STORM Wiki pipeline
def run_storm_wiki(partner):
    command = [
        "python", "examples/storm_examples/run_storm_wiki_gpt.py",
        "--retriever", "zyte",
        "--do-research",
        "--do-generate-outline",
        "--do-generate-article",
        "--do-polish-article"
    ]
    
    # Use subprocess to run the command and capture output
    process = subprocess.Popen(command, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    
    # Send the partner name as input
    stdout, stderr = process.communicate(input=f"A 3 page blog for the company {partner}\n")
    
    # Print the output
    print(f"Output for {partner}:")
    print(stdout)
    if stderr:
        print("Errors:")
        print(stderr)
    print("\n" + "="*50 + "\n")

# Loop through each partner and run the pipeline
for partner in partners:
    run_storm_wiki(partner)
```

```python
import base64
import vertexai
from vertexai.generative_models import GenerativeModel, Part, SafetySetting
from vertexai.preview.prompts import Prompt

text1 = """{ARTICLE_CONTENT}
Create a 2 page markdown article which tells abou the company, from this, Dont write \'\'page 1\' or \'page 2\' make it continuous"""

generation_config = {
    "max_output_tokens": 8192,
    "temperature": 0.1,
    "top_p": 0.95,
}

safety_settings = [
    SafetySetting(
        category=SafetySetting.HarmCategory.HARM_CATEGORY_HATE_SPEECH,
        threshold=SafetySetting.HarmBlockThreshold.OFF
    ),
    SafetySetting(
        category=SafetySetting.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT,
        threshold=SafetySetting.HarmBlockThreshold.OFF
    ),
    SafetySetting(
        category=SafetySetting.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT,
        threshold=SafetySetting.HarmBlockThreshold.OFF
    ),
    SafetySetting(
        category=SafetySetting.HarmCategory.HARM_CATEGORY_HARASSMENT,
        threshold=SafetySetting.HarmBlockThreshold.OFF
    ),
]
```

```python
def generate_article(textvar_1):
    vertexai.init(project="daply-356808", location="us-central1")
    variables = [
        {
            "ARTICLE_CONTENT": [textvar_1],
        },
    ]
    prompt = Prompt(
        prompt_data=[text1],
        model_name="gemini-1.5-flash-002",
        variables=variables,
        generation_config=generation_config,
        safety_settings=safety_settings,
    )
    # Generate content using the assembled prompt. Change the index if you want
    # to use a different set in the variable value list.
    responses = prompt.generate_content(
        contents=prompt.assemble_contents(**prompt.variables[0]),
    )

    return responses.candidates[0].text

# article_data = generate_article("A 3 page blog for the company Zuber Lawler")
```

```python
import os
import markdown2
from weasyprint import HTML
import zipfile

articles_folder = "/home/cosmos/datagen/tools/storm-open-search/results/gpt"
articles_folder_full_path = [os.path.join(articles_folder, article) for article in os.listdir(articles_folder)]

for article in articles_folder_full_path:
    print(f"Generating article for {article}")
    article_path = os.path.join(article, "storm_gen_article.txt")
    with open(article_path, "r") as file:
        article_content = file.read()
        new_article = generate_article(article_content)

    new_article_path = os.path.join(article, "gemini_article.txt")
    with open(new_article_path, "w") as file:
        file.write(new_article)
    
    # Convert the markdown to HTML
    html = markdown2.markdown(new_article)
    
    # Convert HTML to PDF
    pdf_path = os.path.join(article, "gemini_article.pdf")
    HTML(string=html).write_pdf(pdf_path)

# Zip all the PDFs in a folder
output_folder = "/home/cosmos/datagen/tools/storm-open-search/results/pdfs"
os.makedirs(output_folder, exist_ok=True)
zip_path = os.path.join(output_folder, "gemini_articles.zip")

with zipfile.ZipFile(zip_path, 'w') as zipf:
    for article in articles_folder_full_path:
        pdf_path = os.path.join(article, "gemini_article.pdf")
        if os.path.exists(pdf_path):
            zipf.write(pdf_path, os.path.basename(pdf_path))

print(f"PDFs have been zipped and saved to {zip_path}")
```

```python
import markdown
from weasyprint import HTML, CSS
from weasyprint.text.fonts import FontConfiguration
import re

def create_linked_citations(text):
    # Find all citations in the format [n]
    citations = re.findall(r'\[(\d+)\]', text)
    
    # Replace each citation with a linked version
    for citation in citations:
        text = text.replace(f'[{citation}]', f'<a href="#ref{citation}">[{citation}]</a>')
    
    # Add anchors to the sources
    lines = text.split('\n')
    for i, line in enumerate(lines):
        if line.startswith(f"{citation}."):
            lines[i] = f'<a id="ref{citation}"></a>' + line
    
    return '\n'.join(lines)

# Read the Markdown file
with open('/home/cosmos/datagen/tools/storm/results/gpt/History_of_Bitcon/storm_gen_article_polished.md', 'r', encoding='utf-8') as file:
    markdown_text = file.read()

# Create linked citations
linked_markdown = create_linked_citations(markdown_text)

# Convert Markdown to HTML
html = markdown.markdown(linked_markdown, extensions=['extra'])

# Create CSS for styling
css = CSS(string='''
    body { font-family: Arial, sans-serif; }
    h1 { color: #333366; }
    a { color: #0000FF; }
''')

# Configure fonts
font_config = FontConfiguration()

# Create HTML object
html = HTML(string=html)

# Generate PDF
html.write_pdf(
    'output.pdf',
    stylesheets=[css],
    font_config=font_config
)

print("PDF has been generated successfully.")
```
